```python
import sqlite3

conn = sqlite3.connect("C:\\outage_poc\\backend\\outage_poc.db")
cursor = conn.cursor()

cursor.executescript("""
PRAGMA foreign_keys = OFF;
DELETE FROM notification_attempt;
DELETE FROM notification;
DELETE FROM outage_customer_map;
DELETE FROM outage_circuit_map;
DELETE FROM outage_event;
DELETE FROM raw_outage_event;
DELETE FROM customer_service_point;
DELETE FROM customer;
DELETE FROM customer_type;
DELETE FROM service_point;
DELETE FROM channel_master;
DELETE FROM alembic_version;
PRAGMA foreign_keys = ON;
""")

conn.commit()
conn.close()
print("All tables cleared!")
```

```python
from datetime import datetime, timedelta

def check(status, actual_end_time, now, already_sent):
    if (
        status in {"restored", "resolved"}
        or (actual_end_time is not None and actual_end_time <= now)
    ) and not already_sent:
        return "Outage Restored"
    return None


now = datetime.now()
print(now)

# ✅ test cases
assert check("restored", None, now, False) == "Outage Restored"
assert check("resolved", None, now, False) == "Outage Restored"
assert check("ongoing", now - timedelta(minutes=1), now, False) == "Outage Restored"
assert check("ongoing", now + timedelta(minutes=1), now, False) is None
assert check("restored", None, now, True) is None


print("All tests passed ✅")
```
