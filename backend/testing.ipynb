```python
from datetime import datetime, timedelta, UTC
from dataclasses import dataclass

@dataclass
class OutageEvent:
    outage_id: int
    status: str
    start_time: datetime | None
    actual_end_time: datetime | None
    cancellation_flag: bool
```

```python
from app.services import planned_outage_service
```

```python
class LifecycleTestHelper:
    def __init__(self, sent: set[tuple[int, str]] | None = None) -> None:
        self.sent = sent or set()

    def _already_sent(self, outage_id: int, notification_type: str) -> bool:
        return (outage_id, notification_type) in self.sent

    def _advance_notice_was_sent_early_enough(self, outage_id: int, now: datetime) -> bool:
        return self._already_sent(outage_id, "Advance Notice")


def _next_due_notification_type(self: LifecycleTestHelper, outage: OutageEvent, now: datetime) -> str | None:
        if outage.cancellation_flag and not self._already_sent(outage.outage_id, "Cancellation Alert"):
            return "Cancellation Alert"

        status = outage.status.strip().lower()
        actual_end_time = planned_outage_service.ensure_aware_utc(outage.actual_end_time)
        if (
            status in {"restored", "resolved"}
            or (actual_end_time is not None and actual_end_time <= now)
        ) and not self._already_sent(outage.outage_id, "Outage Restored"):
            return "Outage Restored"

        start_time = planned_outage_service.ensure_aware_utc(outage.start_time)
        if start_time is None:
            return None

        if now >= start_time and not self._already_sent(outage.outage_id, "Outage Start"):
            return "Outage Start"

        if status != "scheduled":
            return None

        minutes_until_start = (start_time - now).total_seconds() / 60
        if (
            15 <= minutes_until_start <= 60
            and not self._already_sent(outage.outage_id, "Reminder")
            and self._advance_notice_was_sent_early_enough(outage.outage_id, now)
        ):
            return "Reminder"

        if 0 < minutes_until_start <= 24 * 60 and not self._already_sent(outage.outage_id, "Advance Notice"):
            return "Advance Notice"

        return None
```

```python
now = datetime.now(UTC)

outage = OutageEvent(
    outage_id=1,
    status="Scheduled",
    start_time=now + timedelta(minutes=15),
    actual_end_time=None,

    cancellation_flag=False
)
```

```python
helper = LifecycleTestHelper()
_next_due_notification_type(helper, outage, now)
```
